**Case Description**

An FMCG brand X is available at a large number of stores.
The brand is avaialable as a basic version X and in the form of some value added variants
The price of brand X can vary from store to store depending on the dicounts/promos that the store is offering.
In some stores there are special displays for brand X

**Variable Desciprtion**

Price: Avg price of brand X in that store

Feature: Whether the value added variant of brand X is present in store(1) or not(0)

Display: Whether any store level promo of brand X is present in the store (1) or not(0)

Price-Competitors: Prices of 5 competition brands in the store

**Aim**

The aim of this project is to assess the impact the 4 Ps of marketing has on sales.

Here feature(indicates if the product is value added or not) is a proxy for product

display is a proxy for promotion

**Requirements & Metrics to Evaluate The Model**

Adj r^2>0.6

Mean Absolute Error

**Reading the dataframe**

```python

import pandas as pd

# Read the Excel file
df_train = pd.read_excel('/content/Marketing_Mix Modelling Dataset.xlsx', sheet_name='MM Training data')
df_test = pd.read_excel('/content/Marketing_Mix Modelling Dataset.xlsx', sheet_name='MM Test data')

# Now you have two dataframes: df_train and df_test
# You can proceed with your analysis using these dataframes.
```

**Performing Step wise regression- Price V/S Sales & Evaluating Model Performance With MAE**

```python
import statsmodels.formula.api as sm

# Fit the linear regression model
model = sm.ols('Sales ~ Price', data=df_train).fit()

# Print the summary of the model
print(model.summary())

# Extract the coefficients
intercept = model.params['Intercept']
price_coeff = model.params['Price']

# Create the 'Predicted Sales' column in df_test
df_test['Predicted Sales'] = intercept + df_test['Price'] * price_coeff

# Calculate the Mean Absolute Error (MAE)
mae = (df_test['Sales'] - df_test['Predicted Sales']).abs().mean()

print(f"Mean Absolute Error (MAE): {mae}")


```

**Performing Step wise regression- Price V/S Sales,Feature & Display & Evaluating Model Performance With MAE**

```python
# Fit the linear regression model with multiple independent variables
model = sm.ols('Sales ~ Price + Feature + Display', data=df_train).fit()

# Print the summary of the model
print(model.summary())

# Extract the coefficients
intercept = model.params['Intercept']
price_coeff = model.params['Price']
feature_coeff = model.params['Feature']
display_coeff = model.params['Display']


# Create the 'Predicted Sales' column in df_test
df_test['Predicted Sales'] = intercept + df_test['Price'] * price_coeff + df_test['Feature'] * feature_coeff + df_test['Display'] * display_coeff

# Calculate the Mean Absolute Error (MAE)
mae = (df_test['Sales'] - df_test['Predicted Sales']).abs().mean()

print(f"Mean Absolute Error (MAE): {mae}")
```

**To study the interaction effect**

```python
df_train['PriceDisplay_Interaction'] = df_train['Price'] * df_train['Display']
```

# Understanding Interaction Effects in Marketing Mix Modelling

In real-world marketing scenarios, the impact of one variable often depends on the presence or value of another. This is where interaction effects become important. Rather than modeling variables in isolation, we introduce pairwise interaction terms—products of two variables—to capture these dependencies.

For example, the effect of price on sales might differ depending on whether an in-store display is present. Similarly, the influence of a brand’s price may change depending on the competitor’s pricing, or an advertisement feature may only be effective when combined with a price discount.



**Regression Model With Interaction Effect Of Price & Display(Promotion)**

```python
# Fit the linear regression model with specified independent variables including interaction term
model = sm.ols('Sales ~ Price + Feature + Display + PriceCompetitor1 + PriceCompetitor4 + PriceDisplay_Interaction', data=df_train).fit()

# Print the summary of the model
print(model.summary())

# Extract the coefficients
intercept = model.params['Intercept']
price_coeff = model.params['Price']
feature_coeff = model.params['Feature']
display_coeff = model.params['Display']
price_competitor1_coeff = model.params['PriceCompetitor1']
price_competitor4_coeff = model.params['PriceCompetitor4']
price_display_interaction_coeff = model.params['PriceDisplay_Interaction']


# Create the 'Predicted Sales' column in df_test
df_test['PriceDisplay_Interaction'] = df_test['Price'] * df_test['Display']
df_test['Predicted Sales'] = (intercept + df_test['Price'] * price_coeff + df_test['Feature'] * feature_coeff +
                              df_test['Display'] * display_coeff + df_test['PriceCompetitor1'] * price_competitor1_coeff +
                              df_test['PriceCompetitor4'] * price_competitor4_coeff +
                              df_test['PriceDisplay_Interaction'] * price_display_interaction_coeff)

# Calculate the Mean Absolute Error (MAE)
mae = (df_test['Sales'] - df_test['Predicted Sales']).abs().mean()

print(f"Mean Absolute Error (MAE): {mae}")
```
